```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def derivative(t,y,omega, k):
    """
    theta_dot = phi
    phi_dot = - Omega^2 * sin(theta)
    """
    theta = y[0]
    phi = y[1]
    theta_dot = phi
    phi_dot = - omega**2 * np.sin(theta) - k*phi
    
    
    return np.array([theta_dot, phi_dot])
```

```python
theta_0 = np.pi/2 # initial angle
phi_0 =  1.519011*np.pi # initial speed

t_span = (0,50)
y_0 = (theta_0, phi_0)
t_eval = np.linspace(0,50,1000)

g = 9.81
l = 1
omega = np.sqrt(g/l)

k = 0.4
```

```python
sol = solve_ivp(derivative, t_span, y_0, t_eval=t_eval,args=(omega,k))
```

```python
height = l - np.cos(sol.y[0]) * l 
```

```python
plt.plot(sol.t,height)
```
