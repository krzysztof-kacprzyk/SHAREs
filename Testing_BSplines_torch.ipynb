```python
import torch
from bsplinePyTorchblocks.BSplineLayer import BSplineLayer
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = BSplineLayer(1, 1, n_bases=10, shared_weights=True,bias=True, weighted_sum=False)
input = torch.linspace(-10, 10, 1000).reshape(-1, 1)
output = m(input)
print(output.size())
```

```python
output_np = output.detach().numpy()

t = np.linspace(-10, 10, 1000)
plt.plot(t, output_np[:,0])
```
