```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({'font.size': 20})

# Create a figure and a set of subplots
fig, ax = plt.subplots()
# Increase the space above and below the plot
fig.subplots_adjust(top=0.80, bottom=0.15)

x = np.linspace(-0.99, 0.95, 100)
y = (1+x)/np.sqrt(1 - x**2)
# Set the font size

plt.title(r'$y = \frac{1+x}{\sqrt{1 - x^2}}$', pad=30)
plt.plot(x, y, label=r'$y = \frac{1+x}{\sqrt{1 - x^2}}$', linewidth=4)
plt.xlabel('x')
plt.ylabel('y')


plt.savefig('figures/Figure_3.pdf')
```
